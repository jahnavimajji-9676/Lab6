```python
#A2
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# read data from csv file
data = pd.read_csv('preprocessed_data - preprocessed_data.csv')

# polarity as independent variable and subjectivity as dependent variable
X = data[['polarity']].values
y = data['subjectivity'].values

# Splitting data into training and testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
reg = LinearRegression()
reg.fit(X_train, y_train)
y_pred = reg.predict(X_test)

# Mean squared error
mse = mean_squared_error(y_test, y_pred)
print("Mean Square Error: ", mse)

# Plot the linear regression
plt.scatter(X_test, y_test, color='black')
plt.plot(X_test, y_pred, color='blue', linewidth=3)
plt.xticks(())
plt.yticks(())
plt.show()
```

```python
#A3
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# read data from csv file
data = pd.read_csv('preprocessed_data - preprocessed_data.csv')

# polarity as independent variable and subjectivity as dependent variable
X = data[['polarity']].values
y = data['subjectivity'].values

# Splitting data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
reg = LinearRegression()
reg.fit(X_train, y_train)
y_pred = reg.predict(X_test)

# Mean squared error
mse = mean_squared_error(y_test, y_pred)
print("Mean Square Error: ", mse)

# Plot the linear regression
plt.scatter(X_test, y_test, color='black')
plt.plot(X_test, y_pred, color='blue', linewidth=3)
plt.title('Linear Regression Model')
plt.xlabel('Polarity')
plt.ylabel('Subjectivity')
plt.show()
```

```python
#A4
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_recall_fscore_support
# read data from csv file
data = pd.read_csv('preprocessed_data - preprocessed_data.csv')

# Assuming 'polarity' and 'category' are the features and target for binary classification
X = data[['polarity']].values
y = data['category'].apply(lambda x: 1 if x > 20 else 0).values

# Splitting dataset into a training set and a test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Create a logistic regression classifier
logreg = LogisticRegression()

# Train the classifier with the training data
logreg.fit(X_train, y_train)

# Use the trained model to predict the test set
y_pred = logreg.predict(X_test)

# accuracy
accuracy = accuracy_score(y_test, y_pred)

# precision, recall, F1-score, and support
precision, recall, f1_score, support = precision_recall_fscore_support(y_test, y_pred, average='binary')

print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1_score)
print("Support:", support)
```

```python
#A5
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error

# read data from csv file
data = pd.read_csv('preprocessed_data - preprocessed_data.csv')
# polarity as independent variable and subjectivity as dependent variable
X = data[['polarity']].values
y = data['subjectivity'].values

# Splitting data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Creating a Decision Tree Regressor
regressor_tree = DecisionTreeRegressor()

# Training the model
regressor_tree.fit(X_train, y_train)

# Making predictions
y_pred_tree = regressor_tree.predict(X_test)

# Mean squared error
mse_tree = mean_squared_error(y_test, y_pred_tree)
print("Mean Squared Error (Regression Tree): ", mse_tree)
```

```python
from sklearn.neighbors import KNeighborsRegressor

# Creating a k-NN Regressor
regressor_knn = KNeighborsRegressor(n_neighbors=3)

# Training the model
regressor_knn.fit(X_train, y_train)

# Making predictions
y_pred_knn = regressor_knn.predict(X_test)

# Mean squared error
mse_knn = mean_squared_error(y_test, y_pred_knn)
print("Mean Squared Error (k-NN Regressor): ", mse_knn)
```
